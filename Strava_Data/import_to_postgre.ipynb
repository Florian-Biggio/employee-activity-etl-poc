```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL
import pandas as pd

# Configuration
DB_CONFIG = {
    'host': 'localhost',
    'database': 'strava_poc',
    'user': 'strava_admin',
    'password': 'aqwzsxedc',
    'port': '5432'
}

ACTIVITIES_CSV = 'strava_simulation.csv'

def create_db_connection():
    """Create a database connection"""
    connection_url = URL.create(
        "postgresql+psycopg2",
        username=DB_CONFIG['user'],
        password=DB_CONFIG['password'],
        host=DB_CONFIG['host'],
        port=DB_CONFIG['port'],
        database=DB_CONFIG['database']
    )
    return create_engine(connection_url)
```

```python
import pandas as pd
from sqlalchemy import text

def import_activities_csv(csv_path):
    """Import the activities CSV into the database"""
    engine = create_db_connection()
    
    # Read the CSV file (note it's tab-delimited)
    df = pd.read_csv(csv_path, sep='\t', na_values='')
    if 'ID' in df.columns:
        df = df.drop(columns=['ID'])
    
    with engine.connect() as connection:
        # Create table if it doesn't exist
        connection.execute(text("""
            CREATE TABLE IF NOT EXISTS public.employee_activities (
                ID SERIAL PRIMARY KEY,
                ID_salarie INTEGER NOT NULL,
                Date_de_debut TIMESTAMP NOT NULL,
                Sport_type VARCHAR(50) NOT NULL,
                Distance_m INTEGER,
                Date_de_fin TIMESTAMP NOT NULL,
                Commentaire TEXT
            )
        """))
        connection.commit()
        
        # Insert data
        df.to_sql(
                'employee_activities', 
                con=engine,   # ✅ use engine
                if_exists='append', 
                index=False,
                method='multi'  # ✅ bulk insert for speed
            )
    
    print(f"Successfully imported {len(df)} records")
```

```python
import pandas as pd
from sqlalchemy import text

def export_activities_to_csv(output_csv):
    """Export activities from PostgreSQL to a CSV file"""
    engine = create_db_connection()
    
    with engine.connect() as connection:
        # Read the table into a DataFrame
        df = pd.read_sql('SELECT * FROM public.employee_activities ORDER BY "ID"', connection)

    
    # Save as CSV
    df.to_csv(output_csv, index=False)
    print(f"✅ Exported {len(df)} records to {output_csv}")
    return df

# Example usage:
df_exported = export_activities_to_csv("exported_activities.csv")
print(df_exported.head())
```
